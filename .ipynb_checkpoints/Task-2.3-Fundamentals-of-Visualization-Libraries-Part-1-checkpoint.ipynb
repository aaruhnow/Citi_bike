### 4. Create a column with the count of trips per day and merge it with the weather dataframe

```python
df_temp.columns
```

```python
# Groupby the count of the trips 
df_trip_counts_per_day = pd.DataFrame(df.groupby(['date'])['ride_id'].count()).reset_index()
```

```python
df_trip_counts_per_day.head(100)
```

```python
# Merge the trip counts with the weather DataFrame
df_trip_counts_per_day.rename(columns = {'ride_id':'bike_rides_daily'}, inplace = True)
df = df.merge(df_trip_counts_per_day, on = "date", how = 'outer', indicator = True)
print(df['_merge'].value_counts(dropna = False))
print("Shape of January 1st is", df[df['date'] == '2022-01-01'].shape) # Check 
print("Shape of January 2nd is", df[df['date'] == '2022-01-02'].shape) # Second check 

df_temp = df.set_index('date')

print(df_temp.columns)
```

```python
df_temp.head(1000)
```

### 5. Create a line chart of bike trip counts and temperatures plotted on a dual axis

```python
# Create figure and axis objects with subplots()
fig,ax = plt.subplots(figsize=(10, 5))

# Create a plot of the bike rides
ax.plot(df_temp['bike_rides_daily'],color = "red")

# Create the dual axis
ax2 = ax.twinx()

# Create a plot with the temperatues on the y-axis using second axis object
ax2.plot(df_temp["avgTemp"], color = "blue")
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10, 5))

# Create a plot of the bike rides
line1, = ax.plot(df_temp['bike_rides_daily'], color="navy", label="Bike Rides Daily")
ax.set_xlabel("Year 2022", fontsize=14)  # Set x-axis label
ax.set_ylabel("Bike rides daily", color="navy", fontsize=14)  # Set y-axis label

# Create the dual axis
ax2 = ax.twinx()

# Make a plot with the temperatures on the y-axis using the second axis object
line2, = ax2.plot(df_temp["avgTemp"], color="red", label="Average Temperatures")
ax2.set_ylabel("Average temperatures", color="red", fontsize=14)

# Combine legends from both axes
lines = [line1, line2]
labels = [line.get_label() for line in lines]
ax.legend(lines, labels, loc="upper left", fontsize=12)

# Add title
plt.title("Temperature and trips in 2022", fontsize=18)

# Show the plot
plt.show()
```

### 6. In a markdown cell, briefly explain your code; how did you use Matplotlib to achieve this? Which paradigm did you use?

#### I used Matplotlib's Object-Oriented Approach to create two figures that visualize the relationship between bike rides and average temperatures. This approach offers greater flexibility and control compared to the Procedural (Functional) approach, as it allows for the creation and management of figure and axis objects. This is particularly useful for complex plots, such as those with multiple subplots or dual y-axes. The first plot visualizes both bike rides and average temperatures using a dual y-axis, enabling a clear comparison of the two datasets. The second plot enhances this visualization with descriptive labels and a title, along with customized colors—navy for bike rides and red for temperatures—to distinguish between the datasets more effectively. Using this approach allowed me to create dual-axis plots and manage figure elements such as labels, colors, and titles in a structured manner. The visualization reveals that bike rides tend to increase during warmer temperatures and decrease during colder temperatures, indicating a strong correlation between temperature and bike rentals. This insight suggests that Citi Bike customers are more likely to rent bikes during warmer weather, providing valuable information for planning and resource allocation.
