### 2. Read in data and join it

```python
# Create a list with all files in the folder using a list comprehension
folderpath = r"Data" 
filepath = [os.path.join(folderpath, name) for name in os.listdir(folderpath) if name.endswith(".csv")]
filepath
```

### 3. Get weather data using NOAA's API

```python
# Defining token
token = 'XiCjTQWbEHpVzRSYNbfzbAriDgFyzVjH'
```

```python
# Get the API
r = requests.get('https://www.ncdc.noaa.gov/cdo-web/api/v2/data?datasetid=GHCND&datatypeid=TAVG&limit=1000&stationid=GHCND:USW00014732&startdate=2022-01-01&enddate=2022-12-31', headers={'token':token})
```

```python
# Checking the Content of r.text
print(r.text[:100])
```

```python
# Load the API response as a json
d = json.loads(r.text)
```

```python
# Secure all items in the response that correspond to TAVG
avg_temps = [item for item in d['results'] if item['datatype']=='TAVG']
```

```python
# Get only the date field from all average temperature readings
dates_temp = [item['date'] for item in avg_temps]
```

```python
# Get the temperature from all average temperature readings
temps = [item['value'] for item in avg_temps]
```

```python
# Put the results in a dataframe
df_temp = pd.DataFrame()
```

```python
# Get only date and cast it to date time; convert temperature from tenths of Celsius to Fahrenheit
df_temp['date'] = [datetime.strptime(d, "%Y-%m-%dT%H:%M:%S") for d in dates_temp]
df_temp['avgTemp'] = [(float(v)/10.0) * 9/5 + 32 for v in temps]
```

### 4. Merge the weather data with New York CitiBike data set and export it to a CSV file

```python
# Ensure date columns are consistent before merge
print(df['date'].dtype, df_temp['date'].dtype)  
```

#### Time fields are aligned properly before merge

```python
# Doing a left join and merging df and df_temp and creating df_merged dataframe
df_merged = df.merge(df_temp, how = 'left', on = 'date', indicator = True)
```

```python
df_merged.head()
```

```python
# Count occurrences of each merge case
merge_status = df_merged['_merge'].value_counts(dropna = False)
print(merge_status)
```

```python
# Compare the number of rows before and after the merge
print("Original rows in df:", len(df))
print("Rows after merge:", len(df_merged))
```

```python
# Check for missing values in the new fields from df_temp
missing_values = df_merged[df_merged['_merge'] == 'left_only']
missing_values
```

```python
# Count total NaN values
total_nan = df_merged.isna().sum().sum()
print("Total NaN values:", total_nan)
```
